## Asignación calificada por pares: Segmentación y agrupación de vecindarios en Toronto

# Parte 1



```python
import pandas as pd
```

### Leer la tabla desde la fuente, es decir, la URL de Wikipedia a la Lista de códigos postales de Canadá

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
read_table = pd.read_html(url,header=[0])
df = read_table[0]
df.head(10)
```

### Cambiar el nombre de los encabezados de columna

```python
df = df.rename(index=str, columns={'Postcode':'PostalCode','Neighbourhood':'Neighborhood'})
df.head()
```

### Ignorar las celdas con un municipio que no está asignado.

```python
df = df[df.Borough !='Not assigned']
df.reset_index(drop=True,inplace=True)
df.head()
```

### Agrupamiento de Neighborhood basado en el mismo código postal

```python
df = df.groupby('PostalCode',as_index=False).agg(lambda x: ','.join(set(x.dropna())))
df.head()
```

### Si una celda tiene un distrito pero un vecindario no asignado, entonces el vecindario será el mismo que el distrito

```python
df.loc[df['Neighborhood'] == 'Not assigned','Neighborhood'] = df['Borough']
df.head()
```
